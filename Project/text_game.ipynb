```python
# 초기 상태 설정
player_name = input("당신의 이름은 무엇입니까? ")
hp = 100
inventory = []

print(f"{player_name}님, 일주일간 던전 생존기에 오신 걸 환영합니다!")
print(f"초기 체력: {hp}")
print("")

# 1일차 (월요일)
print("[월요일] 배고픈 아침입니다. 빵을 발견했습니다.")
eat_bread = input("빵을 먹겠습니까? (예/아니오): ")
if eat_bread == "예":
    hp += 10
    inventory.append("빵")
    print("빵을 먹고 힘이 납니다! (체력 +10)")
    print(f"보유 아이템 {inventory}")
else:
    hp -= 10
    print("아침을 거르고 기운이 빠집니다... (체력 -10)")

print(f"현재 체력: {hp}")
if hp <= 0:
    print("😢 체력이 바닥나 쓰러졌습니다. 게임 오버")
else:
    print("")

# 2일차 (화요일)
if hp > 0:
    print("[화요일] 독버섯이 널려 있습니다.")
    mushroom = input("독버섯을 피해가겠습니까? (예/아니오): ")
    if mushroom == "예":
        print("현명한 선택입니다!")
    else:
        hp -= 30
        print("독버섯을 먹고 배탈이 났습니다... (체력 -30)")

    print(f"현재 체력: {hp}")
    if hp <= 0:
        print("😢 체력이 바닥나 쓰러졌습니다. 게임 오버")
    else:
        print("")

# 3일차 (수요일)
if hp > 0:
    print("[수요일] 상자 발견")
    treasure_chest = input("(예/아니오)") # 상자 확인 여부
    if treasure_chest == "예":
        treasure_chest.append("검")
        print("'검'을 발견")
        print(f"보유 아이템 {inventory}")
    else:
        print("상자에서 뭐가 나올지 몰라 무시하고 지나갑니다.")

if hp > 0:
    print("덩굴 발견")
    vine = input("(예/아니오)") # 덩굴 지나갈지 말지
    if vine == "예":
        if "검" in inventory:
            print("검 사용해서 덩굴 제거")
        else:
            hp -= 30
            print("덩굴에 숨겨져 있던 가시에 베임 (체력 -30)")
    else:
        hp -= 10
        print("덩굴 피해가느라 지침 (체력 -10)")

    print(f"현재 체력: {hp}")
    if hp <= 0:
        print("😢 체력이 바닥나 쓰러졌습니다. 게임 오버")
    else:
        print("")

# 4일차 (목요일)
if hp > 0:
    pass
```
