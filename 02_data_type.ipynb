## 2.자료형 (Data Type)
- 자료의 종류와 구조를 정의하는 것

```python
# 문자열(string, str)
my_str1 = '' # 빈문자열
my_str2 = " " # 공백문자열
my_str3 = "안녕하세요" # "요하녕한세" -> 문자열은 '순서가 있는 자료형'

# 문자열 여러줄로 만들기
multi_str = """코
딩
온
"""

print(multi_str)

print(type(multi_str))
```

```python
# 따옴표 속에 따옴표 쓰기
print("'지식'은 우정을 대신하지 않아")
print('"지식"은 우정을 대신하지 않아')
```

```python
# 2. 정수형 (integer, int)
# 파이썬의 정수형의 실질적 크기 제한이 없다
# 변수가 저장되는 메모리의 용량의 한계 = 정수형의 크기 제한
# 대부분의 언어는 크기 제한이 존재

my_int1 = 100

print(type(my_int1))
```

```python
# 3. 실수형 (float) = 떠다니는
# 부동소수점 방식

my_float = 3.14
my_float1 = 100.0

print(type(my_float1))
```

```python
# 4. 논리형 (boolean, bool)
# 참과 거짓을 표현하는 자료형

print(True)
print(False)

print(type(True))
print(type(False))
```
