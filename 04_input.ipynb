## 사용자 입력 (input)
- input() : 콘솔을 통해 사용자로부터 문자열 형태의 입력을 받는 함수
    - input을 통해 받은 입력은 항상 문자열로 입력됨

```python
# input 사용법
# my_input = input()

# input을 통해 받은 입력 -> 문자열
# 숫자로 활용할시 '형변환'을 해줘야 됨

# 정수 계산시 정수형으로 변환
a = int(input())
b = int(input())

print(a + b)
```

```python
# 실수 계산을 하려면 실수형으로 변환

c = float(input())
d = float(input())

print(c + d)
```

```python
# input의 입력에 대한 설명문 추가

name = input("이름을 입력하세요 :")
print(f"제 이름은 {name}입니다.")
```

```python
# 문자열을 쪼개주는 함수 : split()
# split은 구분자를 기준으로 문자열을 쪼개서 리스트를 반환함
# 구분자의 기본값은 공백

과일1, 과일2, 과일3 = "수박 사과 참외".split()
print(과일1)
print(과일2)
print(과일3)

# 여러 자료 입력하기
# input에 적용
과일1, 과일2, 과일3 = input().split()
print(과일1)
print(과일2)
print(과일3)
```

```python
# split() 안에 쪼개는 기본값 변경 가능

과일 = "수박-사과-참외".split(-)
print(과일)
```
